```python
# Gender,Age,Occupation,Student Status,Average Grade (Expect),Average ECTS (Expect),ECTS Summer,ECTS Winter,Grade Summer (AVG),Grade Winter (AVG)


import pandas as pd
import numpy as np
from pathlib import Path
```

```python
file_path = Path('data.csv')
if file_path.exists():
	df = pd.read_csv(file_path)
	print(df)
else:
	print(f"File {file_path} does not exist.")

df
```

```python
df = df.dropna()
df = df.reset_index(drop=True)
df
len(df)
```

```python
df.describe()
```

```python
import dfutil as dfutil
df = dfutil.remove_outliers_based_on_IQR(df)
len(df)
```

```python
import scipy.stats as stats

# Perform the relative t-test
t_stat_grades, p_value_grades = stats.ttest_rel(df['Grade Winter (AVG)'], df['Grade Summer (AVG)'])

print(f"T-statistic: {t_stat_grades}")
print(f"P-value: {p_value_grades}")
```

```python

# Perform the relative t-test on the ECTS columns
t_stat_ects, p_value_ects = stats.ttest_rel(df['ECTS Winter'], df['ECTS Summer'])

print(f"T-statistic: {t_stat_ects}")
print(f"P-value: {p_value_ects}")
```

```python
import matplotlib.pyplot as plt

# Calculate the mean grades for summer and winter
mean_grades = df[['Grade Summer (AVG)', 'Grade Winter (AVG)']].mean()

# Plot the mean grades
mean_grades.plot(kind='bar', color=['orange', 'blue'])
plt.title('Average Grades in Summer vs Winter')
plt.ylabel('Average Grade')
plt.xlabel('Season')
plt.xticks(ticks=[0, 1], labels=['Summer', 'Winter'], rotation=0)
plt.show()
```
