```python
# Gender,Age,Occupation,Student Status,Average Grade (Expect),Average ECTS (Expect),ECTS Summer,ECTS Winter,Grade Summer (AVG),Grade Winter (AVG)


import pandas as pd
import numpy as np
from pathlib import Path
```

```python
file_path = Path('data.csv')
if file_path.exists():
	df = pd.read_csv(file_path)
	print(df)
else:
	print(f"File {file_path} does not exist.")

df
```

```python
import matplotlib.pyplot as plt

# Calculate the mean grades for summer and winter
mean_grades = df[['Grade Summer (AVG)', 'Grade Winter (AVG)']].mean()

# Plot the mean grades
mean_grades.plot(kind='bar', color=['orange', 'blue'])
plt.title('Average Grades in Summer vs Winter')
plt.ylabel('Average Grade')
plt.xlabel('Season')
plt.xticks(ticks=[0, 1], labels=['Summer', 'Winter'], rotation=0)
plt.show()
```
